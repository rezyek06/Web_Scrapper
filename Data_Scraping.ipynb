```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
```

```python
def Web_Scraper(url):
  page = requests.get(url)
  bs = BeautifulSoup(page.text, 'html')

  table = bs.find_all('table')[0]
  column_names = []
  for x in table.find_all('th'):
    column_names.append(x.text.strip())

  ds = pd.DataFrame(columns=column_names)

  data = []
  for x in table.find_all('tr')[1:]:
    row_data = x.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    ds.loc[len(ds.index)] = individual_row_data

  return ds
```

```python
Web_Scraper(url)
```
